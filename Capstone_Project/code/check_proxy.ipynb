```python
import requests

def check_proxy_availability(proxies):
    unavailable_proxies = []
    
    for proxy in proxies:
        print(f"Testing {proxy}")
        try:
            response = requests.get('https://www.google.com', proxies={'http': proxy, 'https': proxy}, timeout=5)
            if response.status_code != 200:
                unavailable_proxies.append(proxy)
                print(f"Proxy {proxy} unavailable")
        except requests.exceptions.RequestException:
            unavailable_proxies.append(proxy)
    
    return unavailable_proxies


unavailable_proxies = check_proxy_availability(proxies)
print("Unavailable Proxies:", unavailable_proxies)
```
