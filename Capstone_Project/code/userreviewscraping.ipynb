```python
# Check if the progress file exists
if os.path.exists("progress_ratings2.txt"):
    with open("progress_ratings2.txt", "r") as file:
        start_index = int(file.read())
else:
    start_index = 0


# Check if the ratings data file exists
if os.path.exists("df_ratings2.csv"):
    df_ratings2 = pd.read_csv("df_ratings2.csv")
    ratings = df_ratings2["ratings"].tolist()
    users = df_ratings2["user_id"].tolist()
else:
    ratings = []
    users = []

new_users = df_ratings2['user_id'].unique()[start_index:] 

break_counter = 0

for i, user in enumerate(new_users, start=start_index):
    print("=" * 80)
    if break_counter == 10:
        print("More than 10 bad requests")
        break

    proxy = random.choice(proxies)
    print(f"Scraping for {user}")
    rating = get_user_ratings_useridinput(user, proxy)
    ratings.append(rating)
    users.append(user)
    time.sleep(random.uniform(0, 2))

    # Save data after processing 5 users
    if i % 5 == 0:
        with open("progress_ratings2.txt", "w") as file:
            file.write(str(i + 1))
        df_ratings2 = pd.DataFrame({"user_id": users, "ratings": ratings})
        df_ratings2.to_csv("df_ratings2.csv", index=False)
        print("df_ratings2 updated")
        time.sleep(random.uniform(1,3))

# Remove the progress file as the processing is complete
os.remove("progress_ratings2.txt")
```
